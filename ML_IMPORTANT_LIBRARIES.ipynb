S**TACKING**

```python
# np.hstack = horizontal stacking means combine two array as horizontal with same size
import numpy as np
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
arr=np.hstack((a1,a2))
print(arr)
```

```python
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
arr=np.vstack((a1,a2))
print(arr)
```
